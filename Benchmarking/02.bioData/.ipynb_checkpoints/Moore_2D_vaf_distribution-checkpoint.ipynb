```python
input_dir = '/data/project/Alzheimer/EM_cluster/Moore_data/Ver3/TwoSample_all_woMosaic/PD28690/adrenal_gland_zona'

# input_file = [i for i in os.listdir(input_dir) if 'txt' in i]
input_file = ['glomerulosa_L1_fasciculata_L1_input.txt', 'glomerulosa_L2_fasciculata_L2_input.txt']
input_file.sort()
for file in input_file:
    A = file.rstrip().split('_')[0]
    B = file.rstrip().split('_')[2]
    
    A_shared_vaf_list = []
    B_shared_vaf_list = []
    
    A_only_vaf_list = []
    B_only_vaf_list = []
    
    A_total_vaf_list = []
    B_total_vaf_list = []

    A_shared_alt_list = []
    B_shared_alt_list = []
    
    A_only_alt_list = []
    B_only_alt_list = []
    
    A_total_alt_list = []
    B_total_alt_list = []
    
    A_shared_depth_list = []
    B_shared_depth_list = []
    
    A_only_depth_list = []
    B_only_depth_list = []
    
    A_total_depth_list = []
    B_total_depth_list = []    
    
    B_only_mutation = []
    B_shared_mutation = []
    
    print(file)
    print(A,B)
    f = open(input_dir + '/' + file, 'r')
    for fline in f:
        if fline == "":
            continue
        ftab = fline.rstrip().split('\t')
        depth_alt = ftab[2]
        A_depth = int(ftab[2].rstrip().split(',')[0])
        A_alt = int(ftab[2].rstrip().split(',')[1])

        B_depth = int(ftab[2].rstrip().split(',')[2])
        B_alt = int(ftab[2].rstrip().split(',')[3]) 

        if A_depth == 0:
            B_only_vaf = B_alt/B_depth
            B_only_vaf_list.append(B_only_vaf)
            B_total_vaf = B_alt/B_depth
            B_total_vaf_list.append(B_total_vaf)
            
            B_only_alt_list.append(B_alt)
            B_total_alt_list.append(B_alt)
            
            B_only_depth_list.append(B_depth)
            B_total_depth_list.append(B_depth) 
            
            B_only_mutation.append(ftab[0])
        elif B_depth == 0:
            A_only_vaf = A_alt/A_depth
            A_only_vaf_list.append(A_only_vaf)
            A_total_vaf = A_alt/A_depth
            A_total_vaf_list.append(A_total_vaf)
            
            A_only_alt_list.append(A_alt)
            A_total_alt_list.append(A_alt)
            
            A_only_depth_list.append(A_depth)
            A_total_depth_list.append(A_depth) 
            
        else:
            B_shared_vaf = B_alt/B_depth
            B_shared_vaf_list.append(B_shared_vaf)            
            A_shared_vaf = A_alt/A_depth
            A_shared_vaf_list.append(A_shared_vaf)                
            B_total_vaf = B_alt/B_depth
            A_total_vaf = A_alt/A_depth
            A_total_vaf_list.append(A_total_vaf)
            B_total_vaf_list.append(B_total_vaf)
            
            B_shared_alt_list.append(B_alt)
            B_total_alt_list.append(B_alt)
            A_shared_alt_list.append(A_alt)
            A_total_alt_list.append(A_alt)
            
            B_shared_depth_list.append(B_depth)
            B_total_depth_list.append(B_depth)
            A_shared_depth_list.append(A_depth)
            A_total_depth_list.append(A_depth)
            
            B_shared_mutation.append(ftab[0])
    f.close()
    print()
    print(A)
    print(len(A_total_vaf_list))
    print(len(A_shared_vaf_list))
    print(len(A_only_vaf_list))
    fig = plt.figure(figsize = (10,5))
    ax = fig.add_subplot()
    plt.tick_params(axis = 'x', labelsize = 15)
    ax.set_xlabel('VAF', fontsize = 20)
    ax.set_title('ZG L1', fontsize=30)
    sc = ax.set_xlim(0, 1.0); ax.set_ylim(0, 10)
    sns.distplot(A_total_vaf_list, hist=False, ax=ax, label=A + '_total', color = 'darkorange')
    sns.distplot(A_shared_vaf_list, hist=False, ax=ax, label=A + '_shared', color = 'blue')
    sns.distplot(A_only_vaf_list, hist=False, ax=ax, label=A + '_only', color = 'forestgreen')
    ax.legend()
    plt.show()
    print(B)
    print(len(B_total_vaf_list))
    print(len(B_shared_vaf_list))
    print(len(B_only_vaf_list))
    fig = plt.figure(figsize = (10,5))
    ax = fig.add_subplot()
    plt.tick_params(axis = 'x', labelsize = 15)
    ax.set_xlabel('VAF', fontsize = 20)
    ax.set_title('ZF L1', fontsize=30)
    sc = ax.set_xlim(0, 1.0); ax.set_ylim(0, 10)
    sns.distplot(B_total_vaf_list, hist=False, ax=ax, label=B + '_total', color = 'darkorange')
    sns.distplot(B_shared_vaf_list, hist=False, ax=ax, label=B + '_shared', color = 'blue')
    sns.distplot(B_only_vaf_list, hist=False, ax=ax, label=B + '_only', color = 'forestgreen')
    ax.legend()
    plt.show()       
    
    print(A)
    fig = plt.figure(figsize = (10,5))
    ax = fig.add_subplot()
    plt.tick_params(axis = 'x', labelsize = 15)
    ax.set_xlabel('ALT', fontsize = 20)
    ax.set_title('ZG L1', fontsize=30)
    sc = ax.set_xlim(0, 30)
    sns.distplot(A_total_alt_list, hist=False, ax=ax, label=A + '_total', color = 'darkorange')
    sns.distplot(A_shared_alt_list, hist=False, ax=ax, label=A + '_shared', color = 'blue')
    sns.distplot(A_only_alt_list, hist=False, ax=ax, label=A + '_only', color = 'forestgreen')
    ax.legend()
    plt.show()
    print(B)
    fig = plt.figure(figsize = (10,5))
    ax = fig.add_subplot()
    plt.tick_params(axis = 'x', labelsize = 15)
    ax.set_xlabel('ALT', fontsize = 20)
    ax.set_title('ZF L1', fontsize=30)
    sc = ax.set_xlim(0, 30)
    sns.distplot(B_total_alt_list, hist=False, ax=ax, label=B + '_total', color = 'darkorange')
    sns.distplot(B_shared_alt_list, hist=False, ax=ax, label=B + '_shared', color = 'blue')
    sns.distplot(B_only_alt_list, hist=False, ax=ax, label=B + '_only', color = 'forestgreen')
    ax.legend()
    plt.show()         
    
    print(A)
    fig = plt.figure(figsize = (10,5))
    ax = fig.add_subplot()
    plt.tick_params(axis = 'x', labelsize = 15)
    ax.set_xlabel('DEPTH', fontsize = 20)
    ax.set_title('ZG L1', fontsize=30)
    sc = ax.set_xlim(0, 100)
    sns.distplot(A_total_depth_list, hist=False, ax=ax, label=A + '_total', color = 'darkorange')
    sns.distplot(A_shared_depth_list, hist=False, ax=ax, label=A + '_shared', color = 'blue')
    sns.distplot(A_only_depth_list, hist=False, ax=ax, label=A + '_only', color = 'forestgreen')
    ax.legend()
    plt.show()
    print(B)
    fig = plt.figure(figsize = (10,5))
    ax = fig.add_subplot()
    plt.tick_params(axis = 'x', labelsize = 15)
    ax.set_xlabel('DEPTH', fontsize = 20)
    ax.set_title('ZF L1', fontsize=30)
    sc = ax.set_xlim(0, 100)
    sns.distplot(B_total_depth_list, hist=False, ax=ax, label=B + '_total', color = 'darkorange')
    sns.distplot(B_shared_depth_list, hist=False, ax=ax, label=B + '_shared', color = 'blue')
    sns.distplot(B_only_depth_list, hist=False, ax=ax, label=B + '_only', color = 'forestgreen')
    ax.legend()
    plt.show()    
    
#     fig, ax = plt.subplots()
#     sns.distplot(A_shared_vaf_list, hist=False, ax=ax, label=A + '_shared', color = 'blue')
#     sns.distplot(A_only_vaf_list, hist=False, ax=ax, label=A + '_only', color = 'darkorange')
#     ax.legend()
#     plt.show()
    
#     fig, ax = plt.subplots()
# #     sns.distplot(A_shared_vaf_list, hist=False, ax=ax, label=A + '_shared', color = 'blue')
#     sns.distplot(B_shared_vaf_list, hist=False, ax=ax, label=B + '_shared', color = 'red')
# #     sns.distplot(A_only_vaf_list, hist=False, ax=ax, label=A + '_only', color = 'darkorange')
#     sns.distplot(B_only_vaf_list, hist=False, ax=ax, label=B + '_only', color = 'forestgreen')
#     ax.legend()
#     plt.show()
#     print()
```

## one sample Depth, Alt, VAF distribution

```python
input_dir = '/data/project/Alzheimer/EM_cluster/Moore_data/Ver2/OneSample_woMosaic/PD28690/adrenal_gland_zona'

input_file = [i for i in os.listdir(input_dir) if 'txt' in i]
input_file.sort()
print(input_file)
region_number = {'glomerulosa':0, 'fasciculata':1, 'reticularis':2}
vaf_arr = np.zeros((3,5))
depth_arr = np.zeros((3,5))
alt_arr = np.zeros((3,5))
variant_count_arr = np.zeros((3,5))

for file in input_file:
    print(file)
    region = file.rstrip().split('_')[0]
    number = file.rstrip().split('_')[1][1:]
    vaf_list = []
    depth_list = []
    alt_list = []
    total_variant = 0
    f = open(input_dir + '/' + file, 'r')
    for fline in f:
        ftab = fline.rstrip().split('\t')
        depth_alt = ftab[2]
        depth = int(ftab[2].rstrip().split(',')[0])
        alt = int(ftab[2].rstrip().split(',')[1])
        vaf = alt/depth
        total_variant += 1
        depth_list.append(depth)
        alt_list.append(alt)
        vaf_list.append(vaf)
        
    print(len(vaf_list))
    fig = plt.figure(figsize = (10,5))
    ax = fig.add_subplot()
    plt.tick_params(axis = 'x', labelsize = 15)
    ax.set_xlabel('VAF', fontsize = 20)
    ax.set_title(file, fontsize=15)
#     sc = ax.set_xlim(0, 1.0); ax.set_ylim(0, 10)
    sns.distplot(vaf_list, hist=False, ax=ax, color = 'darkorange')
    plt.show()
    fig = plt.figure(figsize = (10,5))
    ax = fig.add_subplot()
    plt.tick_params(axis = 'x', labelsize = 15)
    ax.set_xlabel('DEPTH', fontsize = 20)
    ax.set_title(file, fontsize=15)
#     sc = ax.set_xlim(0, 1.0); ax.set_ylim(0, 10)
    sns.distplot(depth_list, hist=False, ax=ax, color = 'blue')
    plt.show()        
    fig = plt.figure(figsize = (10,5))
    ax = fig.add_subplot()
    plt.tick_params(axis = 'x', labelsize = 15)
    ax.set_xlabel('ALT', fontsize = 20)
    ax.set_title(file, fontsize=15)
#     sc = ax.set_xlim(0, 1.0); ax.set_ylim(0, 10)
    sns.distplot(alt_list, hist=False, ax=ax, color = 'forestgreen')
    plt.show()        
    f.close()
    
    median_vaf = get_median(vaf_list)
    median_depth = get_median(depth_list)
    median_alt = get_median(alt_list)
    
    vaf_arr[region_number[region],int(number)-1] = round(median_vaf,2)
    depth_arr[region_number[region],int(number)-1] = round(median_depth,2)
    alt_arr[region_number[region],int(number)-1] = round(median_alt,2)
    variant_count_arr[region_number[region],int(number)-1] = round(total_variant,0)
    print()
```

```python
print(vaf_arr)
print(depth_arr)
print(alt_arr)
print(variant_count_arr)
```
