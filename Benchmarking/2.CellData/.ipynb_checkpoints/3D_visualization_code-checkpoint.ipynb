```python
import sys
import os
import scipy
import gzip
from scipy.stats import binom
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
answer_f = open('/data/project/Alzheimer/YSscript/EM_MRS/data/npvaf/MRS_3D/500_1_0.1_0.1/M1-2_M1-5_M1-7_input/16/npvaf.txt','r')
input_dir = '/data/project/Alzheimer/YSscript/EM_MRS/data/combinedoutput/MRS_3D/500_1_0.1_0.1/M1-2_M1-5_M1-7_input/16'
clement_f = open(input_dir + '/CLEMENT_hard_1st.membership.txt','r')
clement_list = []
for fline in clement_f:
    membership = str(fline.rstrip())

    clement_list.append(membership)
clement_f.close()
print(len(clement_list))

answer_df = pd.DataFrame(columns=['ID','block1','block2','block3','CLUSTER'])
clement_df = pd.DataFrame(columns=['ID','block1','block2','block3','CLUSTER'])


idx = 0
for fline in answer_f:
    ftab = fline.rstrip().split('\t')
    if ftab[0] == '':
        continue
    answer_df.loc[idx] = ftab[0], float(ftab[1]), float(ftab[2]), float(ftab[3]), str(ftab[4])    
    clement_df.loc[idx] = ftab[0], float(ftab[1]), float(ftab[2]), float(ftab[3]), clement_list[idx]

    idx += 1
answer_f.close()
```

```python
sm_ec =  open(input_dir + '/CLEMENT_decision.results.txt','r')
for fline in sm_ec:
    if fline.startswith('NUM_CLONE'):
        print(fline)
    if fline.startswith('NUM_CHILD'):
        print(fline)
    if fline.startswith('score'):
        print(fline)
sm_ec.close()
```

### Answer

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize = (20,20))
ax = fig.add_subplot(111, projection='3d')
plt.tick_params(axis = 'x', labelsize = 15)
plt.tick_params(axis = 'y', labelsize = 15)
plt.tick_params(axis = 'z', labelsize = 15)
ax.set_xlabel('sample 1', fontsize = 20)
ax.set_ylabel('sample 2', fontsize = 20)
ax.set_zlabel('sample 3', fontsize = 20)
ax.set_title('Answer', fontsize=30)

sc = ax.set_xlim3d(0, 0.6); ax.set_ylim3d(0, 0.6); ax.set_zlim3d(0, 0.6)

for group_name, group_idx in answer_df.groupby('CLUSTER').groups.items():
    y = answer_df.iloc[group_idx,2]
    x = answer_df.iloc[group_idx,1]
    z = answer_df.iloc[group_idx,3] 
    
    # color_list = firebrick royalblue darkorange forestgreen mediumorchid
    # royalblue darkorange forestgreen mediumorchid firebrick gold deeppink darkcyan
    
    if group_name == 'S0':
        ax.scatter(x,y,z, label=group_name, alpha = 0.9, s=500,color='firebrick')
    elif group_name == 'V1':
        ax.scatter(x,y,z, label=group_name, alpha = 0.9, s=500,color='royalblue')
    elif group_name == 'V2':
        ax.scatter(x,y,z, label=group_name, alpha = 0.9, s=500,color='darkorange')
    elif group_name == 'S0,V2':
        ax.scatter(x,y,z, label=group_name, alpha = 0.9, s=500,color='forestgreen')
    elif group_name == 'S0,V1':
        ax.scatter(x,y,z, label=group_name, alpha = 0.9, s=500,color='mediumorchid')
    elif group_name == '5':
        ax.scatter(x,y,z, label=group_name, alpha = 0.9, s=500,color='darkcyan')
    elif group_name == 'FP':
        ax.scatter(x,y,z, label=group_name, alpha = 0.9, s=500,color='black')
        
ax.legend(loc = 'upper right', fontsize = 20)
# ax.figure.savefig('/data/project/Alzheimer/EM_cluster/00.RESULT/3D_Answer_phylogeny1.pdf')
```
